** Feature selection from gene expression data: 3000 most variable genes.**

Because we had many more gene expression features than samples, in order to avoid overfitting we wanted to perform feature selection. One approach to this was inspired by the major TCGA paper on breast cancer (Cancer Genome Atlas Network, 2012. Comprehensive molecular portraits of human breast tumours. Nature, 490(7418), pp.61–70.), in which the authors eliminated features that showed very little variance between samples.



```python
## extract the 3000 genes that are most highly variable in TCGA paper

# 1. Change dataframe values from str to float
expr = expr.astype('float')
```

```python
# 2. Count how many NA in each row -- none. Good.
# expr.count(axis=1).describe()

# 3. Eliminate rows in which mean of normalized count is low.
# These are more likely to be subject to random noise.
exprMeans = expr.mean(axis=1)
exprMeansLo = exprMeans[exprMeans < 10]
exprMeansHi = exprMeans[exprMeans >= 10]
exprHi = expr.ix[exprMeansHi.index.values]
# print exprMeansHi[0:5]
# print exprHi.head()
expr = exprHi

# 4. Determine the mean-scaled SD of each gene.
# Scaled to means of so highly expressed genes don't dominate.
# Perhaps log-transform could have accomplished something similar.
exprSDs = exprHi.std(axis=1)
# print exprSDs.head()
exprSDsMN = exprSDs / exprMeansHi
# print exprSDsMN.head()

# 5. Plot variances
fig, ax = plt.subplots()
exprSDsMN.hist(ax=ax, bins=25, bottom=0.1)
ax.set_yscale('log')
plt.title('Distribution of mean-scaled SDs of all genes')
plt.xlabel('Mean-scaled standard deviation')
plt.ylabel('# genes (log scaled)')
plt.show()
# ax.set_xscale('log')

# 6. Sort, draw cutoff
# choosing 3000 because stdev is ~1.5 x mean; somewhat arbitrary
exprSDsMNvar3000= exprSDsMN.order(ascending=False)[0:3000]
print exprSDsMNvar3000.head()
var3000_genes = exprSDsMNvar3000.index.values
# var3000_genes is for subsetting expr later
```

Next we had to select a clinical feature to predict from our genomic data. We decided on neoplasm.diseasestage, which is the stage assigned to the disease by the diagnosing physician. Cancer staging is the process of determining the extent to which a cancer has developed by spreading. Contemporary practice is to assign a number from I-IV to a cancer, with I being an isolated cancer and IV being a cancer which has spread to the limit of what the assessment measures. The stage generally takes into account the size of a tumor, whether it has invaded adjacent organs, how many regional (nearby) lymph nodes it has spread to (if any), and whether it has appeared in more distant locations (metastacized) (http://en.wikipedia.org/wiki/Cancer_staging). Note that cancer stage is assigned at the time of the initial cancer diagnosis, and does not change if the cancer progresses (http://www.cancer.org/treatment/understandingyourdiagnosis/staging).

```python
#DJL Let's see what the disease stages are and how many we have of each

#first define a function to give a nice presentation of
#the unique counts, which we can reuse for any dataframe and feature 
def uniqueCounts(df, feature):
    u = pd.DataFrame(itemfreq(df[feature].values), columns=[feature, 'count'])
    u.set_index(feature, inplace=True)
    u.columns.name = feature
    u.index.name = ''
    return u

neoplasm_diseasestage_df = uniqueCounts(clinpick, 'neoplasm.diseasestage')

# Make a plot
plot = neoplasm_diseasestage_df.plot(kind='bar', legend = False)
plt.title('Number of patients within each disease stage')
plt.xlabel('Disease stage')
plt.ylabel('Number of patients')
plt.show()

neoplasm_diseasestage_df
```

**Exploratory analysis with SVD**

In order to tell whether the data separated naturally into groups and whether those groups correlated with our cancer stage classes, we tried two different ways to do SVD that were shown in lecture and homeworks.

```python
# SVD v1: using scipy.linalg

import scipy.linalg

def standardize(X):
    return (X - np.mean(X, axis=0)) / np.std(X, axis=0)

X_trainStandard = standardize(X_train)
samples,genes = X_trainStandard.shape
U,s,Vh = scipy.linalg.svd(X_train)
S = scipy.linalg.diagsvd(s, samples, genes)# blow up s
PC1 = np.dot(U, S[:,:1])
PC2 = np.dot(U, S[:,1:2])
```

```python
# SVD v1
# plot the cumulative variance explained by number of components
variances = np.diagonal(S*S/np.sum(S*S))
plt.plot(np.cumsum(variances[0:50]))
plt.title('Cumulative variance explained')
plt.xlabel('# of principal components')
plt.ylabel('Ratio of variance explained')
```

```python
# SVD v1

# Make a scatter plot of the first two principal components
# where Y value is indicated by color

plt.scatter(PC1, PC2, c=Y_train, cmap='brg')
plt.title('First two principal components of expression data')
plt.xlabel('1st PC')
plt.ylabel('2nd PC')
```

```python
# SVD v2: using sklearn.decomposition.TruncatedSVD

from sklearn.decomposition import TruncatedSVD 

svd = TruncatedSVD(n_components=100)
Xsvd = svd.fit_transform(X_trainStandard)#(normalize(X_train))
```

```python
# SVD v2
# plot the cumulative variance explained by number of components
pd.DataFrame(np.cumsum(svd.explained_variance_ratio_)).plot(figsize=(10, 6))
plt.title('Cumulative variance explained')
plt.xlabel('# of principal components')
plt.ylabel('Ratio of variance explained')
```

```python
# SVD v2
# Make a scatter plot of the first two principal components
# where Y value is indicated by color
plt.scatter(Xsvd[:,0],Xsvd[:,1],c=Y_train, cmap='brg')
plt.title('First two principal components of expression data')
plt.xlabel('1st PC')
plt.ylabel('2nd PC')
```

SVD results: From the scatterplot of the first two principal components of the expression data, we found no clear clustering of the data by disease stage. However this did not dissuade us from going ahead and trying to predict disease stage from gene expression because a) the first two principal components do not account for a large fraction of variance, and b) it is possible that there is a small but statistically significant correlation between expression and disease stage not immediately visible in the scatterplot.

Also, given more time, we could color code the PC scatterplot using other clinical features to see if any of them might be a better candidate for prediction than disease stage.

**Description of mutation types (from wikipedia):**
* frame shift deletion/insertion- A frameshift mutation (also called a framing error or a reading frame shift) is a genetic mutation caused by indels (insertions or deletions) of a number of nucleotides in a DNA sequence that is not divisible by three. Due to the triplet nature of gene expression by codons, the insertion or deletion can change the reading frame (the grouping of the codons), resulting in a completely different translation from the original.

* in-frame deletion- Deletion that does not cause a frameshift.

* in-frame insertion- Insertion that does not cause a frameshift.

* missence mutation- a point mutation in which a single nucleotide change results in a codon that codes for a different amino acid

* nonsense mutation- a point mutation in a sequence of DNA that results in a premature stop codon, or a nonsense codon in the transcribed mRNA, and in a truncated, incomplete, and usually nonfunctional protein product.

* nonstop mutation- single base-pair substitutions that occur within translational termination (stop) codons and which can lead to the continued and inappropriate translation of the mRNA into the 3'-untranslated region.

* RNA - not quite sure what is meant by this

* silent- Silent mutations can occur in non-coding regions (outside of genes within introns), or they may occur within exons. When they occur within exons they either do not result in a change to the amino acid sequence of a protein (i.e. a synonymous substitution), or result in the insertion of an alternative amino acid with similar properties to that of the original amino acid, and in either case there is no significant change in phenotype.

* splice site- A splice site mutation is a genetic mutation that inserts, deletes or changes a number of nucleotides in the specific site at which splicing of an intron takes place during the processing of precursor messenger RNA into mature messenger RNA.

```python
from sklearn.cross_validation import StratifiedShuffleSplit

def stratified_train_test_split (X, Y, test_size, random_state=23):
    """
    Similar to train_test_split, but ensures representative class 
    representation in training and test sets.
    """
    split = StratifiedShuffleSplit(Y, n_iter=1, test_size=test_size, random_state=random_state)
    for train_index, test_index in split:
        len(train_index), len(test_index)  #Insert "Print" before this to actually see the lengths
    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    Y_train = Y.iloc[train_index]
    Y_test = Y.iloc[test_index]
    return X_train, X_test, Y_train, Y_test

X_train, X_test, Y_train, Y_test = stratified_train_test_split (X, Y, 0.2)

print X_train.shape
print X_test.shape
print Y_train.shape
print Y_test.shape
print np.mean(Y_train) #fraction of training set with >0 lymph nodes
print np.mean(Y_test) #fraction of test set with >0 lymph nodes
```

Our findings were not far from those of Hajiloo et al. (below), who showed a model with accuracy 60% for predicting cancer from SNP data. Note that this is a slightly different problem, however, since they were predicting future cancer occurrence, and we were predicting cancer stage among people who did have cancer.
